# K-Means Clustering
K-Means clustering is a method of clustering data points together so that the data points belong to the cluster with the nearest average. For a set of $k$ clusters, $\mathbf{S} = {S_1, S_2, ..., S_k}$, and the observations $\mathbf{x}_1, ... \mathbf{x}_n$, the objective is to find

\begin{align}
\underset{\mathbf{S}}{\mathrm{arg min}} \sum_{i=1}^k \sum_{\mathbf{x} \in S_i} \| \mathbf{x} - \mathbf{\mu}_i \|^2.
\end{align}
Source: [Wikipedia](https://en.wikipedia.org/wiki/K-means_clustering)

The [scikit learn](scikit-learn.org) library provides a python interface to a K-Means algorithm, [sklearn.cluster.KMeans](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.k_means.html#sklearn.cluster.k_means).

```python
import numpy as np
import matplotlib.pyplot as plt
# create clusters
C1 = 6*np.random.random_sample((50,2)) + np.array([5., 5.])
C2 = 6*np.random.random_sample((50,2)) + np.array([5., 10.])
C3 = 6*np.random.random_sample((50,2)) + np.array([10., 10.])

# plot clusters
fig, ax = plt.subplots(1,1)
ax.scatter(C1[:,0], C1[:,1], label="C1")
ax.scatter(C2[:,0], C2[:,1], label="C2")
ax.scatter(C3[:,0], C3[:,1], label="C3")
ax.legend()
ax.set_title("Original Clusters")


C = np.concatenate([C1, C2, C3])
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3).fit(C)

K1 = C[kmeans.labels_ == 0]
K2 = C[kmeans.labels_ == 1]
K3 = C[kmeans.labels_ == 2]

# plot clusters
fig, ax = plt.subplots(1,1)
ax.scatter(K1[:,0], K1[:,1], label="K1")
ax.scatter(K2[:,0], K2[:,1], label="K2")
ax.scatter(K3[:,0], K3[:,1], label="K3")
ax.legend()
ax.set_title("KMeans Clusters")
pass
```
